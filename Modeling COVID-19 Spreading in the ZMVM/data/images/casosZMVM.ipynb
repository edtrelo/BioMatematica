```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib.dates as mdates
```

```python
url_casos = "https://raw.githubusercontent.com/edtrelo/BioMatematica/main/Modeling%20COVID-19%20Spreading%20in%20the%20ZMVM/data/cleandata/Casos_Diarios_ZMVM_Confirmados.csv"
```

```python
casos_por_mun = pd.read_csv(url_casos, encoding = 'utf-8', sep = ',', index_col = 1)
```

```python
casos_ZMVM = casos_por_mun.sum(axis = 0) # suma por filas
```

```python
casos_ZMVM = casos_ZMVM[1: -2] # casos del 26 de febrero de 2020 al 25 de febrero de 2023
```

```python
dates = pd.to_datetime(casos_ZMVM.index, infer_datetime_format=True)
```

```python
fig, ax = plt.subplots()
plt.style.use('ggplot')

ax.plot(dates, casos_ZMVM)

ax.set_ylim(-100, max(casos_ZMVM)*1.1)

ax.xaxis.set_major_formatter(mdates.DateFormatter('%b%Y'))
fig.set_size_inches(12, 8)
plt.title("Casos nuevos de COVID-19 confirmados (por da) en la ZMVM", fontsize = 18)
plt.show()
```
